```python
# import all the required libraries

import pandas as pd
```

```python
#extract the excel file which contains the marked required data 

exl = pd.ExcelFile('Data for archana.xlsx', low_memory=False)
dfa = exl.parse(exl.sheet_names[0])
```

```python
# remove all the columns which are not marked and create a dataframe 

dfa.dropna(inplace = True)
feature_columns = dfa['Long Name'].unique()
feature_columns = list(feature_columns) ## because feature_columns is a numpy_array
feature_columns = ['company'] + ['price_date'] + feature_columns
```

```python
xl05 = pd.ExcelFile('2005.xlsx', low_memory=False)
df05 = xl05.parse(xl05.sheet_names[0])
df05[feature_columns]
```

```python
##Remove all the missing data

print("Before removing nullvalues: ",len(df05))
df05.dropna(inplace = True)
print("After removing null",len(df05))
```
