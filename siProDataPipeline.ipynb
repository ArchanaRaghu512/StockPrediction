## Selecting Features 

Only a selected set of features from a total of 122 features are required of this analysis. The list of the features selected is available in 'Data for archana.xlsx'. The following script create a list of features required.

```python
featureFile = pd.ExcelFile('Data for archana.xlsx')      ## select 'Data for archana.xlsx' and create a file object
features = featureFile.parse(featureFile.sheet_names[0]) ## select the first sheet in that file object
features = features.loc[features['Unnamed: 2'] == 'x']   ## Look out for column 'Unnamed: 2'and select features only with 
                                                         ##'X' Marked. 
features = features['short_name']                        ## of the selected features -  select only their shortname
                                                         ## Convert the series into a list and lowercase
features = features.tolist()
features = [fea.lower() for fea in features]
required_features = ['company'] + features               ## create a required_features with 'company' and features selected.
```

```python
required_features = ['company'] + features
df1 = df1[required_features]                             ## adding company to the required features list
```

### Elimination of Null values

```python
print("Before removing nullvalues: ",len(df1))
df1.dropna(inplace = True)
print("After removm"len(df1))
```
